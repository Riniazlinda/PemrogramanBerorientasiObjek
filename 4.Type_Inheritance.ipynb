=======Overriding=====
-Adalah :Suatu kelas turunan bisa memiliki fungsi dengan nama yanag sama seperti fungsi yang ada pada kelas induknya 
-Teknik overriding mengizinkan kelas turunan untuk memiliki fungsi dengan nama yang sama seperti fungsi pada kelas induknya dengan statement yang berbeda di dalamnya.

```python
#OVERRIDING
class Fish: 
    def __init__(self, first_name, last_name="Fish", skeleton="bone", eyelids=False): 
        self.first_name = first_name 
        self.last_name = last_name 
        self.skeleton = skeleton 
        self.eyelids = eyelids 
    def swim(self): 
        print("The fish is swimming.") 
    def swim_backwards(self): 
        print("The fish can swim backwards.") 

class Shark(Fish): 
    def __init__(self, first_name, last_name="Shark", skeleton="cartilage", eyelids=True): 
        self.first_name = first_name 
        self.last_name = last_name 
        self.skeleton = skeleton 
        self.eyelids = eyelids 
    def swim_backwards(self): 
        print("The shark cannot swim backwards, but can sink backwards.") 
ikan=Fish("hiu")
ikan.swim()
ikan.swim_backwards()
ik=Shark("paus")
ik.swim_backwards()
```

====Multiple Inheritance======
-Konsepnya : Memampukan suatu kelas turunan untuk mewarisi sifat-sifat atau atribut-atribut dari kelas-kelas basisnya.

```python
#MULTIPLE INHERITANCE
class Coral: 
    def community(self): 
        print("Coral lives in a community.") 

class Anemone: 
    def protect_clownfish(self): 
        print("The anemone is protecting the clownfish.") 

class CoralReef(Coral, Anemone): 
    pass 
great_barrier = CoralReef() 
great_barrier.community() 
great_barrier.protect_clownfish() 
```

```python
#CONTOH 
class Murid:
    
    def __init__(self):
        self.nama = input("Nama: ")
        
    def display(self):
        print("Nama: ",self.nama)
   
class NilaiPelajaran:
 
    def __init__(self):
        print("Nilai Pelajaran")
        self.math = int(input("Math: "))
        self.biology = int(input("Biology: "))
   
    def display(self):
        print("Rata2 Nilai: ", (self.math + self.biology)/2 )
        


class student(Murid, NilaiPelajaran):
    def __init__(self):
        Murid.__init__(self)
        NilaiPelajaran.__init__(self)
        
    def result(self):
        Murid.display(self)
        NilaiPelajaran.display(self)
        

stu1 = student()
stu1.result()
```

```python
#No.1
#Overriding
#konsep Superlass dan Overriding untuk luas permukaan
class Lingkaran:
    def __init__(self,jari):
        self.j=jari

    def tampilkan(self):
        print("jari-jari:",self.j)
    def luas(self):
        print("Luas=",3.14*(self.j**2))
    def keliling(self):
        print("keliling=",2*3.14*self.j)


class Tabung(Lingkaran):
    def __init__(self,tinggi,jari):
        self.t=tinggi
        super().__init__(jari)
    
    def tampilkanTinggi(self):
        print("Tinggi tabung:",self.t)

    def luas(self):
        print("Luas Tabung:",2*3.14*self.j*self.t)
        

        
l=Lingkaran(6)
l.luas()
l.keliling()
t=Tabung(4,6)
t.tampilkanTinggi()
t.luas()
```

```python
#kelas penempatan yang menampilkan nama pegawai dengan penempatan lokasi kerjanya di toko apa?
class Pegawai:
    def __init__ (self):
        self.nama=input("Nama: ")
        self.nip=input("NIP: ")
    def display(self):
        print("Nama: ",self.nama)
        print("NIP: ",self.nip)

class Toko:
    def __init__(self):
        self.namatoko=input("Nama Toko: ")
        self.lokasi=input("Lokasi: ")
        
    def location(self):
        print("Nama Toko: ",self.namatoko)
        print("Lokasi: ",self.lokasi)
class Penempatan(Pegawai,Toko):
    def __init__(self):
        Pegawai.__init__(self)
        Toko.__init__(self)
    def tampil(self):
        Pegawai.display(self)
        Toko.location(self)

p1=Penempatan()

p1.tampil()
```

```python
#program buat tampilan seperti KRS yang menampilkan 1 mahasiswa dengan beberapa mata kuliah yang diambil dan nilai ipsnya
class Mahasiswa:
    def __init__(self,nama,npm):
        self.nama=nama
        self.npm=npm
    def tampilkan(self):
        print("nama:",self.nama)
        print("nip:",self.npm)
class Matakuliah:
    def __init__(self):
        n=1
        self.kodematkul=[]
        self.matakuliah=[]
        while n <=2:
            a=input("==>Masukkan kode mata kuliah :")
            b=input("==>Masukkan nama mata kuliah :")
            self.kodematkul.append(a)
            self.matakuliah.append(b)
            n+=1
    def tampilkan(self):
        print("Mata kuliah yang diambil :")
        for n in range (len(self.kodematkul)):
            print(self.matakuliah[n],"(",self.kodematkul[n],")")
class Pengambilan(Mahasiswa,Matakuliah):
    def __init__(self,nama,npm):
        Mahasiswa.__init__(self,nama,npm)
        Matakuliah.__init__(self)
        self.nilai=[]
        for n in range (len(self.kodematkul)):
            a=int(input("Masukkan Nilai matakuliah :"))
            self.nilai.append(a)
    def tampilkan(self):
        Mahasiswa.tampilkan(self)
        print("Matakuliah yang diambil :")
        for n in range (len(self.kodematkul)):
            print(self.matakuliah[n],"(",self.kodematkul[n],") Nilai",self.nilai[n])
            jumlah=0
            for n in self.nilai:
                jumlah=jumlah+n
                print("IPS",jumlah/len(self.nilai))
nama= input("masukkan nama:")
npm = int(input("Masukkan NIM :"))
r=Pengambilan(nama,npm)
r.tampilkan()
```
