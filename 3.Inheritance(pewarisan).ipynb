INHERITANCE(PEWARISAN)
Konsep pewarisan : memampukan suatu kelas mewarisi sifat-sifat(termasuk atribut-atributnya) dari kelas induknya ke kelas turunannya
-Kelas Induk : Kelas induk atau kelas basis/parent
-Kelas Turunan : Kelas turunan dari induk kelas/subclass
-Hubungan antar kelas : Kelas Parent tidak bisa membaca sifat-sifat subclass, tetapi subclass yang bisa membaca sifat-sifat  dari parent karena parentlah yang berperan mewarisi sifat-sifatnya ke subclass
-Format yang digunakan dalam pewarisan :
class Kelas_Induk:
    variabel
    def nama_fungsi():
    ....
class Kelas_Turunan(Kelas_Induk):
    variabel
    ...
    def nama_fungsi():
    ....

```python
class User: 
    nama = ""
    def __init__(self, nama): 
        self.nm = nama  
    def printName(self): 
        print ("Name :" , self.nm)  

class Programmer(User):  
    def __init__(self, nama): 
        self.nm = nama   
    def doPython(self): 
        print ("Programming Python")   
brian = User("brian") 
brian.printName()   
diana = Programmer("Diana") 
diana.printName()
diana.doPython() 
```

```python
 class Fish: 
    def __init__(self, first_name, last_name="Fish", skeleton="bone", eyelids=False): 
        self.first_name = first_name 
        self.last_name = last_name 
        self.skeleton = skeleton 
        self.eyelids = eyelids 
    def swim(self): 
        print("The fish is swimming.") 
    def swim_backwards(self): 
        print("The fish can swim backwards.") 


class Trout(Fish):
    pass 

class Clownfish(Fish): 
    def live_with_anemone(self): 
        print("The clownfish is coexisting with sea anemone.") 

terry = Trout("Terry") 
print(terry.first_name + " " + terry.last_name) 
print(terry.skeleton) 
print(terry.eyelids) 
terry.swim() 
terry.swim_backwards() 

casey = Clownfish("Casey") 
print(casey.first_name + " " + casey.last_name) 
casey.swim() 
casey.live_with_anemone() 
```
